Now we are in condition to define a function for preprocessing of documents before we can form terms-document table

````python
# Define preprocessing function
def preprocess(text):
    lemetize=nltk.WordNetLemmatizer().lemmatize
    stem=nltk.PorterStemmer().stem

    tokens=nltk.tokenize.wordpunct_tokenize(text)
    
    punctuations=string.punctuation
    tokens=list(filter((lambda x:x not in punctuations),tokens))
    
    tokens=[lemetize(word) for word in tokens]    
    tokens=[stem(word) for word in tokens]

    stopwords=nltk.corpus.stopwords.words('english')
    stopwords=set(stopwords)

    tokens=list(filter(lambda x:x not in stopwords,tokens))

    return tokens
````

## Now we will form documtnts-term frequency table

````python
from collections import Counter, defaultdict
````

````python
term_fequency=defaultdict(Counter)

for i,text in enumerate(documents):
    tokens=preprocess(text)
    for word in tokens:
        term_fequency[word][i]+=1
print(term_fequency)
````

````python
term_document_matrix=None
term_document_matrix=pd.DataFrame(term_fequency,columns=term_fequency.keys())
term_document_matrix=term_document_matrix.fillna(0)
term_document_matrix
````

````python

df=None
df=pd.DataFrame(columns=['word'])
for x in term_fequency:
    y=term_fequency[x]
    y['word']=x
    df=df.append(y,ignore_index=True)

df=df.fillna(0)
df[[0,1,2,3,4]]=df[[0,1,2,3,4]].astype(int)
for x in df.values:
    print(x)

term_document_matrix
````
